#Plotting image

```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from mpl_toolkits.basemap import Basemap
import netCDF4 as nc
import os

from matplotlib.collections import LineCollection
from matplotlib import cm
import shapefile
```

```
l = os.listdir("/content/data")
```

```
l
```

```
l.sort()
#l.remove(l[0])
l
```

```
final_list = []
for i in l:
  temp_l = []
  date = i[20:28]
  for j in l:
    if date in j:
      temp_l.append(j)
  final_list.append(temp_l)
  l = l[len(temp_l):]
```

```
complete_list = []
for i in final_list:
  if i != []:
    complete_list.append(i)
```

```
complete_list
len(complete_list)
```

```
def data_process(filename):
  df = nc.Dataset(filename, mode = 'r')
  lon = df.groups['PRODUCT'].variables['longitude'][:][0,:,:]
  lat = df.groups['PRODUCT'].variables['latitude'][:][0,:,:]
  data = df.groups['PRODUCT'].variables['nitrogendioxide_tropospheric_column'][:][0,:,:]
  units = df.groups['PRODUCT'].variables['nitrogendioxide_tropospheric_column'].units
  return lon, lat, data, units
```

```
def create_img(filename, date_list): # Get some parameters for the Stereographic Projection
  lon, lat, data, units = [], [], [], []
  lon_0, lat_0 = [], []
  for i in range(len(filename)):
    lon_temp, lat_temp, data_temp, units_temp = data_process(filename[i])
    lon.append(lon_temp)
    lat.append(lat_temp)
    data.append(data_temp)
    units.append(units_temp)
    lon_0.append(lon_temp.mean())
    lat_0.append(lat_temp.mean())

  date = filename[0][20:28]
  yr = date[0:4]
  mon = date[4:6]
  day = date[6:8]
  date = yr + "-" + mon + "-" + day
  date_list.append(date)
  m = Basemap(width=2700000,height=3200000,
            resolution='l',projection='stere', 
            lat_ts = 50,lat_0 = 22,lon_0 = 80)#stere
  xi, yi = [], []
  for i in range(len(filename)):
    xi_temp, yi_temp = m(lon[i], lat[i])
    xi.append(xi_temp)
    yi.append(yi_temp)

  #m.drawparallels(np.arange(-90., 120., 30.), labels=[1,0,0,0], fontsize=10)
  #m.drawmeridians(np.arange(-180., 180., 45.), labels=[0,0,0,1], fontsize=10)

  for i in range(len(filename)):
    m.scatter(xi[i], yi[i], c = data[i], cmap = plt.cm.jet,
            norm = LogNorm(),edgecolors=None, linewidth=0) 

  # Add Coastlines, States, and Country Boundaries
  m.drawcoastlines(linewidth = 0.3)
  #m.drawstates()
  m.drawcountries(linewidth = 0.3)

  # Add Colorbar
  #cbar = m.colorbar(location='bottom', pad="10%")

  #cbar.set_label(units)
  

  # Add Title
  #plt.title('NO2 Total Vertical Columns')
  fig = plt.gcf()
  pngfile = "{0}.png".format(date)
  #plt.show()
  fig.savefig("/content/NO2_Jul-Aug/"+ pngfile, quality = 95, dpi = 300)
  plt.close()
  return date_list
```

```
date_list = []
for i in range(30, 61):
  date_list = create_img(complete_list[i], date_list)
```
